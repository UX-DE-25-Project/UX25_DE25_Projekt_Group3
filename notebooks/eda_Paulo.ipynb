```python
import pandas as pd
import duckdb
import matplotlib.pyplot as plt
```

```python
query = "SELECT * FROM read_json_auto('../src/data/bostader.json')"
df = duckdb.query(query).to_df()
```

```python
print("=" * 40)
print("EDA - Bostadsdata")
print("=" * 40)

print(f"\nAntal bostäder: {len(df)}")
print(f"Städer: {df['stad'].nunique()} st")
print(f"Typ: {df['typ'].value_counts().to_dict()}")
print(f"Upplåtelse: {df['upplåtelseform'].value_counts().to_dict()}")
```

```python
print("\n--- BOYTA ---")
print(f"Minst: {df['boyta'].min()} kvm")
print(f"Störst: {df['boyta'].max()} kvm")
print(f"Medel: {df['boyta'].mean():.0f} kvm")
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
df['stad'].value_counts().plot(kind='bar')
plt.title('Bostäder per stad')
plt.xlabel('Stad')
plt.ylabel('Antal')
```

```python
plt.subplot(1, 2, 2)
df['pris'].hist(bins=30, edgecolor='black')
plt.title('Prisfördelning')
plt.xlabel('Pris (kr)')
plt.ylabel('Antal')

plt.tight_layout()
plt.show()
```
